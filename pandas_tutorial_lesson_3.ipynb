### Accessing Data with `loc` and `iloc`

loc can be used to access specific rows and columns using labels or a boolean array. For example, we can access the first row of our dataframe by doing the following:

```python
coffee.loc[0]
```

loc is inclusive of the upper index, as we can see in the following:

```python
coffee.loc[5:12] # Using slicing
```

```python
coffee.loc[:, ["Day", "Units Sold"]]
```

We can use a boolean condition too. For example, with the following code, we can display only those rows where more than 20 units were sold:

```python
coffee.loc[coffee["Units Sold"] > 20]
```

iloc is used to select data by its integer position. It's helpful when you want to access rows or columns based on their numerical index, much like how you would access elements in a list or array.

```python
coffee.iloc[0] # Selects first row
```

```python
coffee.iloc[:, 0] # Selects first column
```

Note that iloc is not inclusive of the upper index. We select the rows from index 2 up to BUT NOT INCLUDING 5, and only the columns at indexes 0 and 2

```python
coffee.iloc[2:5, [0,2]]
```

#### Understanding the DataFrame Index

We can see that our dataframe has the default integer index. We have 14 rows, going from 0 to 13, with a step of 1.

```python
coffee.index # Shows the index of the DataFrame
```

We can access specific columns by name in different ways:

```python
coffee.Day
```

```python
coffee["Units Sold"]
```

#### Overwriting the Default Index

We can overwrite the default indexing of our dataframe. First I make a deep copy, for demonstrative purposes:

```python
coffee_copy = coffee.copy(deep=True)
```

Let's set the 'Day' column as the new index.

```python
coffee_copy.set_index('Day', inplace=True)
```

```python
coffee_copy.head()
```

With 'Day' as the index, `loc` can now directly access rows using the day names. For example, let's try to get all entries for 'Monday'.

```python
coffee_copy.loc['Monday']
```

```python
coffee_copy.loc['Monday':'Thursday', 'Units Sold']
```

iloc still relies on integer positions:

```python
coffee_copy.iloc[0,:]
```

```python
coffee_copy.iloc[:, :]
```

##### `iloc` Behavior with a Custom Index

If you want 'Day' to be treated as a regular column again so you can select it with iloc, you would need to reset_index() on your DataFrame. Let's create a new DataFrame coffee_reset for this demonstration:

```python
coffee_reset = coffee_copy.reset_index()
print(coffee_reset.head())
```

'Day' is a regular column again:

```python
coffee_reset.iloc[:, :]
```

```python
coffee_reset.iloc[:, 0] # Selects the 'Day' column
```

### Modifying Data in a DataFrame

We can also modify individual or several cells:

```python
coffee.loc[2, "Units Sold"] = 100
```

```python
coffee.head()
```

```python
coffee.loc[3:5, "Units Sold"] = 15
```

```python
coffee.head(7)
```

We can also get individual cells more efficiently using at and iat. We can't use them to get multiple values, though.

```python
coffee.at[2, "Units Sold"]
```

```python
coffee.iat[2,2]
```

## Sorting Values

We can also sort our values. We can provide column names. By default, values will be ascending:

```python
coffee.sort_values(by="Units Sold")
```

ascending = 0 is equivalent to ascending = False

```python
coffee.sort_values(by=["Units Sold", "Coffee Type"], ascending=[0,1])
```

### Iterating Through Rows (`iterrows`)

We can also iterate row by row with a for loop, but this is not recommended as we will lose some of the memory/performance benefits of pandas.

```python
for idx, row in coffee.iterrows():
  if idx > 5:
    break
  print(idx, row, "\n")
```
