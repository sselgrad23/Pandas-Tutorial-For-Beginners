# `.shift()`, `.rank()`, `.cumsum()`, and `.rolling()`

```python
import pandas as pd
import numpy as np
```

```python
coffee = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/warmup-data/coffee.csv')
coffee['price'] = np.where(coffee['Coffee Type'] == 'Espresso', 3.99, 5.99)
coffee['revenue'] = coffee['Units Sold'] * coffee['price']
latte = coffee[coffee['Coffee Type'] == 'Latte'].copy()
coffee.head()
```

Let's compare today's revenue with yesterday's revenue for both coffee types. We can do that using `shift()`:

```python
coffee['yesterday_revenue'] = coffee['revenue'].shift(2)
```

```python
coffee
```

We can find the percent change between yesterday and today's revenue:

```python
coffee['pct_change'] = coffee['revenue'] / coffee['yesterday_revenue'] * 100
coffee
```

Let's go to our bios dataset and rank the athletes by height:

```python
bios = pd.read_csv('https://github.com/KeithGalli/complete-pandas-tutorial/raw/refs/heads/master/data/bios.csv')
bios.head()
```

```python
bios['height_rank'] = bios['height_cm'].rank(ascending=False)
bios.sort_values(['height_rank'])
```

Let's go back to our coffee dataframe. We can calculate cumulative revenue using `.cumsum()`:

```python
coffee['cumulative_revenue'] = coffee['revenue'].cumsum()
coffee
```

We can get a rolling sum of the number of lattes sold over the past 3 days:

```python
latte['3day'] = latte['Units Sold'].rolling(3).sum()
latte
```
