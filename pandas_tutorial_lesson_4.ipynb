# Filtering Data

```python
import pandas as pd
```

```python
olympics_data_sheets = pd.read_excel('https://github.com/KeithGalli/complete-pandas-tutorial/raw/refs/heads/master/data/olympics-data.xlsx', sheet_name=['bios', 'results'])
```

```python
bios = olympics_data_sheets['bios']
bios.head()
```

```python
bios.info()
```

```python
bios.loc[bios['height_cm'] > 220] # Filtering out athletes greater than 220cm tall
```

```python
bios.loc[bios['height_cm'] > 215, ['name', 'height_cm']].head(5) # Also only grabbing name and height
```

```python
# Alternative syntax
bios[bios['height_cm'] > 215][['name', 'height_cm']].head(5)
```

Use & and round brackets to add additional conditions:

```python
bios[(bios['height_cm'] > 215) & (bios['born_country'] == 'USA')]
```

Filtering using string operations. By default, str.contains is case-sensitive. This can be toggled with case=False.

```python
bios[bios['name'].str.contains("sophie", case=False)]
```

Regex syntax is also allowed. | means or. Regex can be turned off with regex=False.

```python
bios[bios['name'].str.contains("sophie|samantha", case=False)]
```

You can also use .isin to only keep rows where a certain column is one of the specified values:

```python
bios[bios['born_country'].isin(['USA', 'Canada']) & (bios['name'].str.startswith("Samantha"))]
```

query can be used to filter rows with less clunky syntax:

```python
bios.query('born_country == "GBR" and NOC == "France"')
```
