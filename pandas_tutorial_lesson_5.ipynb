# Adding and Removing Columns



```python
import pandas as pd
import numpy as np
```

```python
coffee = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/warmup-data/coffee.csv')
```

We can easily add new columns:

```python
coffee['price'] = 4.99
coffee.head()
```

```python
coffee['Coffee Type'].nunique()
```

```python
coffee['Coffee Type'].unique()
```

We can also incorporate numpy. For example, we use np.where to set the new price of Espressos to 3.99 and the new price of all other coffee types to 5.99:

```python
coffee['new_price'] = np.where(coffee['Coffee Type'] == 'Espresso', 3.99, 5.99)
coffee.head()
```

We can also drop individual rows:

```python
coffee_drop_0 = coffee.drop(0, axis=0)
```

```python
coffee_drop_0.head()
```

```python
drop_price = coffee.drop(columns=['price'])
```

```python
drop_price.head()
```

If we actually want to modify the dataframe, we need to set inplace=True:

```python
# coffee.drop(columns=['price'], inplace=True)
# coffee.head()
```

Alternatively, we can also do df = df.drop(...) without inplace=True:

```python
#coffee = coffee.drop(columns=['price'])
#coffee.head()
```

A different way to drop columns is to just specify the columns we do want to keep. This is not recommended if we have a lot of columns we want to keep.

```python
coffee = coffee[['Day', 'Coffee Type', 'Units Sold', 'new_price']]
```

```python
coffee.head()
```

## Other Operations

We can also multiply columns or perform similar kinds of operations:

```python
coffee['revenue'] = coffee['Units Sold'] * coffee['new_price']
coffee.head()
```

We can rename columns, too. Remember to use inplace=True or assignment to ensure your changes persist:

```python
coffee.rename(columns={'new_price': 'price'}, inplace=True)
coffee.head()
```

We can use string operations to create new columns, too:

```python
olympics_bios = pd.read_excel('https://github.com/KeithGalli/complete-pandas-tutorial/raw/refs/heads/master/data/olympics-data.xlsx', sheet_name=['bios'])
bios_new = olympics_bios['bios'].copy() # We do not necessarily need to use deep_copy
```

```python
bios_new['first_name'] = bios_new['name'].str.split(' ').str[0]
bios_new['last_name'] = bios_new['name'].str.split(' ').str[1]
bios_new.head()
```

```python
bios_new.query('first_name =="Sophie"').head()
```

It is often recommended to turn dates into datetime objects. This can be done using pd.to_datetime. You can also set specific date formats.

```python
bios_new['born_datetime'] = pd.to_datetime(bios_new['born_date'], format="%Y-%m-%d")
bios_new.head()
```

```python
bios_new.info()
```

We can now extract the year from our new column of datetime objects using dt.year. By default the years will be floats.

```python
bios_new['born_year'] = bios_new['born_datetime'].dt.year
bios_new[['name', 'born_year']].head()
```

```python
bios_new.info()
```

We can save our new csv with to_csv. We can set index=False to avoid saving it with the additional index column.

```python
bios_new.to_csv('bios_new.csv', index=False)
```

## Applying custom functions

```python
olympics_bios = olympics_bios['bios']
```

We can write lambda functions and apply them:

```python
olympics_bios['height_category'] = olympics_bios['height_cm'].apply(lambda x: 'Short' if x < 165 else ('Average' if x < 185 else 'Tall'))
```

```python
olympics_bios
```

We can also define regular functions and apply them.

axis=0 means we apply the function on the column axis, but axis=1 means we apply it on the row axis.

```python
def categorise_athlete(row):
  if row['height_cm'] < 175 and row['weight_kg'] < 70:
    return 'Lightweight'
  elif row['height_cm'] < 185 or row['weight_kg'] <= 85:
    return 'Middleweight'
  else:
    return 'Heavyweight'

olympics_bios['Category'] = olympics_bios.apply(categorise_athlete, axis=1)
```
