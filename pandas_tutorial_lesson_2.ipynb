# Loading Dataframes From Different File Formats



```python
import pandas as pd
```

All datasets used here are courtesy of Keith Galli: https://github.com/KeithGalli/complete-pandas-tutorial

Pandas can read CSV files from URLs, which is useful for Google Colab. If you have the CSV file stored locally, e.g. in './warmup_data/coffee.csv',
then you can use pd.read_csv('./warmup_data/coffee.csv')

```python
# Courtesy of Keith Galli
coffee = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/warmup-data/coffee.csv')
```

```python
coffee.head() # First five rows of coffee dataframe
```

CSV is a popular file format for storing data as they can be easily read by humans. However, CSV files can take up a lot of storage. Other popular file formats like .feather and .parquet are generally more efficient in terms of storage and performance.

Below, we read a feather file:

```python
feather_results = pd.read_feather('https://github.com/KeithGalli/complete-pandas-tutorial/raw/refs/heads/master/data/results.feather')
```

```python
feather_results.head()
```

Reading a parquet file:

```python
parquet_results = pd.read_parquet('https://github.com/KeithGalli/complete-pandas-tutorial/raw/refs/heads/master/data/results.parquet')
```

```python
parquet_results.head()
```

We can even read Excel spreadsheets (.xlsx files), as shown below.

However, .xlsx files are also not ideal for large datasets, especially when it comes to storage size and read/write performance compared to binary formats like Feather or Parquet.

They are designed for a broader range of functionality, including formatting, formulas, and multiple sheets. This adds overhead not present in the previously seen simpler file formats.

```python
olympics_data = pd.read_excel('https://github.com/KeithGalli/complete-pandas-tutorial/raw/refs/heads/master/data/olympics-data.xlsx')
```

```python
olympics_data.head()
```

This spreadsheet has two sheets, bios and results. By default, pandas will read the first sheet, bios. If we want to read specific sheets, then we have to do the following:

```python
olympics_data_sheets = pd.read_excel('https://github.com/KeithGalli/complete-pandas-tutorial/raw/refs/heads/master/data/olympics-data.xlsx', sheet_name=['bios', 'results'])
```

```python
olympics_data_sheets.keys() # This tells us which sheets we have
```

```python
olympics_data_sheets['bios'].head()
```

```python
olympics_data_sheets['results'].head()
```

We can also convert dataframes into different file formats.

```python
bios = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/master/data/bios.csv')
```

```python
bios.to_csv('bios.csv')
```

```python
bios.to_excel('bios.xlsx')
```

```python
bios.to_feather('bios.feather')
```

```python
bios.to_parquet('bios.parquet')
```
