# Pandas Basics for Beginners

```python
import pandas as pd
```

For the sake of this tutorial, we manually create a small dataframe and fill it up with values. Typically, we would load some other file first and create a dataframe from it using pandas.

```python
df = pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]], columns = ['a','b','c'], index = ['x', 'y', 'z'])
```

```python
df.head() # Outputs first 5 rows
```

```python
df.head(1) # First row
```

```python
df.tail(2) # Last 2 rows
```

```python
print(df.columns)
print(df.index)
print(df.index.to_list())
```

```python
df.info() # Info about the dataframe itself
```

```python
df.describe() # Gives meaningful stats about our data
```

```python
df.nunique() # Number of unique values in each column
```

```python
df['a'].unique() # Number of unique values in column a
```

```python
df.shape # Shape of dataframe
```

```python
df.size # Total number of items
```
