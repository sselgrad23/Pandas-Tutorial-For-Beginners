# Merging and Concatenating Data

```python
import pandas as pd
```

```python
olympics_bios = pd.read_excel('https://github.com/KeithGalli/complete-pandas-tutorial/raw/refs/heads/master/data/olympics-data.xlsx', sheet_name=['bios'])['bios']
```

```python
nocs = pd.read_csv('https://github.com/KeithGalli/complete-pandas-tutorial/raw/refs/heads/master/data/noc_regions.csv')
```

```python
nocs.head()
```

```python
olympics_bios.head()
```

Check documentation for the different ways to merge two datasets.

*   `left_on`: The column from the left DataFrame to use as a key.
*   `right_on`: The column from the right DataFrame to use as a key.
*   Different types of joins: `inner` (default), `left`, `right`, `outer`.

By default, if two columns in two dataframes have the same column name, the column of the left dataframe will receive the `_x` suffix and the right dataframe will receive the `_y` suffix. You can provide custom suffixes using `suffixes=[...]`.

```python
bios = pd.merge(olympics_bios, nocs, left_on='born_country', right_on='NOC', how='left', suffixes=["bios", "nocs"])
bios.head()
```

```python
bios.rename(columns={'region': 'born_country_full'}, inplace=True) # renaming the columns
bios.head()
```

Here is how we can extract the rows where NOCbios and born_country_full aren't the same:

```python
bios[bios['NOCbios'] != bios['born_country_full']][['name', 'NOCbios', 'born_country_full']]
```

We can extract athletes from certain countries in the following manner:

```python
usa = bios[bios['born_country'] == 'USA'].copy()
gbr = bios[bios['born_country'] == 'GBR'].copy()
```

```python
usa.head()
```

```python
gbr.head()
```

To concatenate these two dataframes:

```python
usa_gbr = pd.concat([usa, gbr])
```

```python
usa_gbr.tail()
```

Finally, let's merge both the bios and the results of our Olympics dataset together. The `on='athlete_id'` parameter specifies that the merge should be performed using the common column 'athlete_id' present in both DataFrames. This ensures that the results are joined correctly for each unique athlete.

```python
results = pd.read_csv('https://github.com/KeithGalli/complete-pandas-tutorial/raw/refs/heads/master/data/results.csv')
combined = pd.merge(bios, results, on='athlete_id', how='left')
```
